```python
# Создайте Series из списка: [150, 180, 120, 200, 175] — это калории, сожжённые за 5 дней тренировок.
# Выведите Series на экран.

import pandas as pd

series = pd.Series([150, 180, 120, 200, 175])

print(series)

```

```python
# Создайте Series из списка [150, 180, 120, 200, 175] с собственными индексами: ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'].
# Выведите Series на экран.

import pandas as pd

series = pd.Series([150, 180, 120, 200, 175], index=['Mon', 'Tue', 'Wed', 'Thu', 'Fri'])

print(series)
```

```python
# Используя Series из задания S2:
# Выведите значение за 'Wed' с помощью loc
# Выведите значение на позиции 3 (четвёртый элемент) с помощью iloc

import pandas as pd

series = pd.Series([150, 180, 120, 200, 175], index=['Mon', 'Tue', 'Wed', 'Thu', 'Fri'])

print(f'series.loc: {series.loc['Wed']}')
print(f'series.iloc: {series.iloc[3]}')
```

```python
# Используя Series из задания S2:
# Измените значение за 'Mon' с 150 на 200
# Выведите обновлённую Series целиком

import pandas as pd

series = pd.Series([150, 180, 120, 200, 175], index=['Mon', 'Tue', 'Wed', 'Thu', 'Fri'])

series.loc['Mon'] = 200

print(series)
```

```python
# Создайте Series из словаря:
# calories = {'Пн': 150, 'Вт': 180, 'Ср': 120, 'Чт': 200, 'Пт': 175}
# С помощью фильтрации (маски) выведите только те дни, где калории больше или равны 170.

import pandas as pd

calories = pd.Series({'Пн': 150, 'Вт': 180, 'Ср': 120, 'Чт': 200, 'Пт': 175})

print(calories[calories >= 170])
```

```python
# Создайте Series из списка температур за неделю:
# [22.5, 24.0, 19.5, 21.0, 23.5, 25.0, 26.5]
# Индексы: ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
# Выполните три действия:
# Найдите самую высокую температуру (метод .max())
# Найдите самую низкую температуру (метод .min())
# Найдите среднюю температуру (метод .mean())

import pandas as pd

series = pd.Series([22.5, 24.0, 19.5, 21.0, 23.5, 25.0, 26.5], index=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])

print(f'Самая высокая температура: {series.max()}')
print(f'Самая низкая температура: {series.min()}')
print(f'Средняя температура: {series.mean()}')
```

```python
# Создайте Series из списка оценок ученика:
# [85, 90, 78, 92, 88, 84, 91]
# Индексы: ['Math', 'Science', 'History', 'Art', 'PE', 'Music', 'English']
# Выведите только те предметы, у которых оценка выше 85.

import pandas as pd

series = pd.Series([85, 90, 78, 92, 88, 84, 91], index = ['Math', 'Science', 'History', 'Art', 'PE', 'Music', 'English'])

print(f'Предметы, у которых оценка выше 85: \n{series[series > 85]}')
```
