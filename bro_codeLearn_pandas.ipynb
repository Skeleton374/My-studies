SERIES

```python
# Series

import pandas as pd # импортирование библиотеки и присванивания алиаса

data = [100, 102, 104] # create python list

series = pd.Series(data) # Series с большой буквы из-за того, что это конструктор, а не функция.

print(series)
```

Все наши данные расположены в одном столбце. У нас появился одномерный массив данных - Series. В левой стороне находятся метки, которые начинаются с нуля. Внизу приведены метаданные об этой серии. Тип данных, которые мы храним - 64-битные целые числа

```python
# Series

import pandas as pd # импортирование библиотеки и присванивания алиаса

data = [100.1, 102.2, 104.3] # create python list

series = pd.Series(data) # Series с большой буквы из-за того, что это конструктор, а не функция.

print(series)
```

При изменении типа данных поменялись и метаданные. Теперь это 64-битные вещественные числа.

```python
# Series

import pandas as pd 

data = [100, 102, 104] 

series = pd.Series(data, index =['apartment #1', 'apartment #2', 'apartment #3']) # Обозначение собственного индекса. параметр index ожидает получить один объект (список, массив, кортеж и т.д.)

print(series)
```

```python
# Series

import pandas as pd 

data = [100, 102, 104] 

series = pd.Series(data, index =['apartment #1', 'apartment #2', 'apartment #3']) 

print(series.loc['apartment #1']) # задача: получить прямой доступ к значению внутри серии. loc[(индекс объекта)]
```

```python
# Series

import pandas as pd 

data = [100, 102, 104] 

series = pd.Series(data, index =['apartment #1', 'apartment #2', 'apartment #3']) 

series.loc['apartment #1'] = 200 # этой командой мы изменили структуру самого одномерного массива

print(series) 
```

Прежние данные:

apartment #1    100
apartment #2    102
apartment #3    104
dtype: int64

```python
# Series

import pandas as pd 

data = [100, 102, 104] 

series = pd.Series(data, index =['apartment #1', 'apartment #2', 'apartment #3']) 

print(series.iloc[0]) # определение местоположения по целочисленной позиции
```

iloc[0] всегда работает с целочисленной позицией элемента независимо от того, какие названия у индексов. В данном случае элемент 'apartment #1' находится на позиции 0, а так как iloc[0] обращается именно к нулевой позиции, то они совпадают и ошибки не возникает — iloc просто берет первый по порядку элемент, игнорируя его строковую метку.

Таким образом, мы можем использовать доступ по целочисленному местоположению или доступ по метке.

```python
# Series

import pandas as pd 

data = [100, 102, 104, 200, 202] # добавляем 2 элемента в одномерный список

series = pd.Series(data, index =['a', 'b', 'c', 'd', 'e']) # соотвественно, добавляем индексирование, иначе возникает ошибка

print(series[series >= 200]) # сортировка.
```

```python
# DataFrame 

import pandas as pd 

data = {
    'Name': ['Spongebob', 'Patrick', 'Squidward'], 
    'Age': [30, 35, 50],
} 

df = pd.DataFrame(data, index = ['Employee 1', 'Employee 2', 'Employee 3']) 

# Add a new column
df['Job'] = ['Cook', 'N/A', 'Cashier']

# Add a new row
new_row = pd.DataFrame([{'Name': 'Sandy', 'Age': '28', 'Job': 'Engineer'}],
                       index = ['Employee 4'])

df = pd.concat([df, new_row])

print(df)
```

```python
# DataFrame 

import pandas as pd

data = {
    'Name': ['Spongebob', 'Patrick', 'Squidward'], 
    'Age': [30, 35, 50],
} 

df = pd.DataFrame(data, index = ['Employee 1', 'Employee 2', 'Employee 3']) 

df['Job'] = ['Cook', 'N/A', 'Cashier']

# Add a new rows
new_row = pd.DataFrame([{'Name': 'Sandy', 'Age': 28, 'Job': 'Engineer'},
                        {'Name': 'Eugene', 'Age': 60, 'Job': 'Manager'}],
                       index = ['Employee 4', 'Employee 5'])

df = pd.concat([df, new_row])

print(df)
```
